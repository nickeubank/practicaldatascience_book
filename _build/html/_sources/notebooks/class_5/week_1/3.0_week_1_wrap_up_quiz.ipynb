## Part 1: Conceptual questions

### Q1. What is the primary purpose of data visualization?

A. To create complex and intricate plots
B. To make data more understandable and clearly communicate patterns
C. To make the plot visually appealing, regardless of clarity
D. To show as many variables as possible in one chart

(Correct answer: B)

### Q2. Which type of plot is generally best for showing the relationship between two continuous variables (e.g. height and weight)?

A. Histogram
B. Bar chart
C. Scatter plot
D. Pie chart

(Correct answer: C)

### Q3. What is the primary reason that it is important to label axes on a plot?

A. It makes the plot look more professional
B. It ensures the reader knows what each axis represents
C. Itâ€™s optional and can be skipped to simplify the plot
D. It prevents the need for a legend

(Correct answer: B)

### Q4. Which of the following statements best describes the role of a legend in a plot?

A. It is decorative and can be used to add colors
B. It is only useful if there are more than two series in the data
C. It explains symbols, colors, or lines to clarify the data for the reader
D. It is necessary for all plots, regardless of content

(Correct answer: C)

### Q5. What should be the top consideration when selecting a color palette for a plot?

A. Only colors that match corporate branding
B. Colors that are easily distinguishable and accessible to all readers, including colorblind readers
C. Colors that match the data points
D. The brightest colors available for visibility

(Correct answer: B)

### Q6. In bar blots, where should I set the baseline for the bar?

A. At the value 0.
B. At the minimum value in the dataset being plotted.
C. At the maximum value in the dataset being plotted.
D. Wherever it best highlights the point you are trying to make.

(Correct answer: A)

### Q7. In which of the following is it easiest for humans to perceive differences between data?

A. Color hue
B. Volume
C. Position along a scale
D. Angle
E. Length
F. Area

(Correct answer: C)
