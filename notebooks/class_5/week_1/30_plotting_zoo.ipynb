# Plotting Zoo: 10 ways to visualize the same set of data

One of the strengths and challenges of plotting data in Python with matplotlib is the sheer quantity of ways that you can plot something. Let's show that in action for a simple dataset about the characteristics of cars. For each one, they receive a rating across four categories: speed, cost, cargo capacity, and sustainability, with 1 being the worst and 5 being the best.

```python
import pandas as pd

vehicles = pd.DataFrame.from_dict({
    'Vehicle':['Sedan', 'Pickup Truck', 'Bicycle', '16-wheeler Truck', 'Lamborghini'],
    'Speed':[4, 4, 1, 3, 5],
    'Cost': [3, 3, 5, 2, 1],
    'Cargo': [2, 3, 1, 5, 2],
    'Sustainability': [4, 3, 5, 1, 2],   
})
vehicles
```

```python
# Optional like of code for higher-resolution figures
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np

attributes = ['Speed','Cost','Cargo','Sustainability']
```
