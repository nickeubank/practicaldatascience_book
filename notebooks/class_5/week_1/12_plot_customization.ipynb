# Customizing plot aesthetics

Let's start with our plot from the last example:

```python
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt

# Create noise-free data
maxval = 2*np.pi
x_model = np.linspace(0,maxval,1000)
f = lambda x: np.sin(x)*x
y_model = f(x_model)

# Create noisy data
x_data = np.random.rand(25)*maxval
y_data = f(x_data) + np.random.randn(*x_data.shape)*0.5

# Optional command to make the figures high resolution
%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots(figsize=(7,7))
ax.plot(x_model,y_model, label="Our Model")
ax.scatter(x_data,y_data,color='black', label="Sample Data")
ax.set_xlabel('x [degrees]')
ax.set_ylabel('y')
ax.set_xticks([0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi])
ax.set_xticklabels([0,90,180,270,360])
ax.set_yticks([-6,-4,-2,0,2,4])
ax.set_xlim([0,2*np.pi])
ax.grid(True)
ax.legend()
ax.set_title('Model performance compared to sample data')
ax.spines[['right','top']].set_visible(False)
plt.show()
```

# Customize colors

## Plot colors

Often we want to make sure that our plots stand out. There are numerous [named colors in matplotlib](https://matplotlib.org/stable/gallery/color/named_colors.html) that you can use to color lines, markers, bars, and text, but really, you can pick any color you want. You're able to input colors in matplotlib using RGB triplets (e.g. [8, 153, 46]) or, what I recommend for ease of use, a hexadecimal code (e.g. #08992e). You can use a [colorpicker](https://g.co/kgs/uqsE5Ai) tool to interactively select the color you want to use and then copy the hex code for it. You can use that code like any other color. Let's try that in our example:

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(x_model,y_model, label="Our Model", color='#08992e')
ax.scatter(x_data,y_data, label="Sample Data", color="#d1d1d1")
ax.set_xlabel('x [degrees]')
ax.set_ylabel('y')
ax.set_xticks([0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi])
ax.set_xticklabels([0,90,180,270,360])
ax.set_yticks([-6,-4,-2,0,2,4])
ax.set_xlim([0,2*np.pi])
ax.grid(True)
ax.legend()
ax.set_title('Model performance compared to sample data')
ax.spines[['right','top']].set_visible(False)
plt.show()
```

It can sometimes be cleaner code to separate the act of plotting content from the style and formatting of the plot. For example, let's separate the content creation from the formatting and styling for the above plot by using the `set` method to set our style parameters after we have created the content:

```python
fig, ax = plt.subplots(figsize=(7,7))
my_line = ax.plot(x_model, y_model, label="Our Model", )
my_scatter = ax.scatter(x_data, y_data, label="Sample Data")
ax.set_xlabel('x [degrees]')
ax.set_ylabel('y')
ax.set_xticks([0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi])
ax.set_xticklabels([0,90,180,270,360])
ax.set_yticks([-6,-4,-2,0,2,4])
ax.set_xlim([0,2*np.pi])
ax.grid(True)
ax.set_title('Model performance compared to sample data')
ax.spines[['right','top']].set_visible(False)

plt.setp(my_line,'color','#08992e')
plt.setp(my_scatter,'color','#d1d1d1')
ax.legend()

plt.show()
```
